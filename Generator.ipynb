# Offload with Generator

```python
from cosim import FlowGen, Manager
```

* Initialize a new flow generator with seed for reproducebility

```python


# define a new flow generator with predefined flow in `flowdir`
G = FlowGen(
    "ue", 
    datadir="__infra__/ue/data", 
    flowdir="__flowdb__",
    seed=12,
)
G
```

* Generate a new flow and send it to offloader
* the offloader will decide where to place the tasks

```python
choosen, choosen_inputs, status_code, reponse_json, url = G.generate_flow(
    offloaderURL="http://127.0.0.1:9696"
)
print(status_code)
```
